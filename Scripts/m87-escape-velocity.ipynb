```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import sys
G = 6.67e-11
M = (6.0e12)*1.99e30
def velocity_behavior(v_0,r):
    sigma = 591. #km/s
    R_s = 32000
    r_0 = 120
    v_r = np.sqrt(v_0**2 - (sigma**2)*(np.log(R_s**2 + r**2) - np.log(R_s**2 + r_0**2)))
    return v_r
"""
for i,f in enumerate(v_r):
    if np.isnan(v_r[i]):
        v_r[i] = 0
    else:
        None
"""

ve = np.sqrt(2*G*M/(1000000*3.08e16))
print ve/1000
```

```python
r = np.linspace(10,50000000,10000)
vr_2800 = velocity_behavior(2800,r)
vr_2500 = velocity_behavior(2500,r)
vr_3000 = velocity_behavior(3100,r)
vr_3300 = velocity_behavior(3250,r)
file_name = 'velocity_behavior.png'
plt.figure(figsize=(7,7))
params = {'axes.labelsize': 18,
          'axes.titlesize':20, 
          'font.size': 14, 
          'legend.fontsize': 14, 
          'xtick.labelsize': 14, 
          'ytick.labelsize': 14
         }
matplotlib.rcParams.update(params)
ax = plt.gca()
ax.minorticks_on()
ax.tick_params(which='minor',direction='in')
ax.tick_params(size = 6, width = 1, direction='in')
ax.grid(alpha=0.25)
ax.tick_params(direction='in')
plt.semilogx(r, vr_2500,label='$v_{i}$ = 2500 km/s')
plt.semilogx(r, vr_2800, label='$v_{i}$ = 2800 km/s')
plt.semilogx(r, vr_3000,label='$v_{i}$ = 3100 km/s')
plt.semilogx(r, vr_3300,label='$v_{i}$ = 3250 km/s')
plt.axhline(y=2300,color='red',linestyle='--',linewidth=1.25,label='Observed Velocity \n 2300 km/s')
plt.axvline(x=3000000,color='black',linestyle='--',linewidth=1.25,label='Estimated Distance')
plt.title('Velocity Behavior of an Object Within \nthe M87 Dark Matter Halo')
plt.xlabel('Radial Distance From M87 (pc)')
plt.ylabel('Radial Velocity (km/s)')
plt.legend()
save_file = './Images/'+file_name
plt.savefig(save_file)
plt.show()
```
